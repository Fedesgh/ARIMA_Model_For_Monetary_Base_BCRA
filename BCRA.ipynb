### ARIMA Model for "held_by_public(1)"

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base_change = basem24["DAILYCHANGE"]
base_stock = basem24["DAILYSTOCK"]
```

```python
base_change.describe()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(base_stock["held_by_public(1)"])
```

At first sight we can see that it is not stationary, then we will use differencing

```python
dif1= base_stock["held_by_public(1)"].astype(int) - base_stock["held_by_public(1)"].astype(int).shift(1)
dif1.dropna(inplace=True)
dif1 = dif1.astype(int)
dif1.plot()
```

```python
#Defining Train set
dif1_train = dif1.loc[:"2024-04-01"]
```

```python

from statsmodels.tsa.stattools import adfuller

test_result = adfuller(dif1_train.values)
print("p value:", test_result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

acf = plot_acf(dif1.values)
```

```python
pacf = plot_pacf(dif1.values)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(dif1.values.astype(int) ,order=([2,8],0,[4,5,8]))
model_fit = model.fit()
model_fit.summary()
```

```python

pred = pd.DataFrame( model_fit.predict(start=len(dif1_train) , end=len(dif1), dynamic=True))
pred.rename(columns={0:"prediction"} , inplace=True)
pred
```

```python
#Creating the prediction column

import numpy as np
x = np.full(len(dif1_train)-1 , np.nan)
x = pd.Series(x)
prediction = pd.DataFrame({"prediction" : x} )
prediction = pd.concat([prediction , pred])

prediction = prediction.set_index(dif1.index)
```

```python

dif1_copy = dif1.to_frame()
dif1_copy["prediction"] = prediction
```

```python
plt.figure(figsize=(24,12))
sns.lineplot(dif1_copy.loc["2024-04":])
```
